```python
import pandas as pd
import akshare as ak
from pyecharts.charts import Bar
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
def get_sw3_detail():
  sw2 = ak.sw_index_second_info()
  sw3 = ak.sw_index_third_info()
  sw3['一级行业'] = sw3['上级行业'].apply(lambda x: sw2.set_index('行业名称')['上级行业'][x])
  return sw3
```

```python
indu_counts = get_sw3_detail().groupby('一级行业')['成份个数'].sum().sort_values()
```

```python
indu_top10 = indu_counts.tail(10)
```

```python
bar = (
    Bar()
    .add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
    .add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
)
bar.load_javascript()
```

```python
bar = (
    Bar()
    .add_yaxis('test', indu_top10.values.tolist())
    .add_xaxis(indu_top10.index.tolist())
    .reversal_axis()
)
bar.render_notebook()
```
